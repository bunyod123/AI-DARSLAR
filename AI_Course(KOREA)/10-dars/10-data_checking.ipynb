```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, LabelEncoder
robastScaling = RobustScaler()
minmaxScaling = MinMaxScaler()
standartScaling = StandardScaler()
labelEncoding = LabelEncoder()
```

```python
df = pd.read_csv(r"C:\Users\bunyo\OneDrive\Desktop\git_dataset\medical_insurance.csv")
```

```python
df.info()
```

```python
cols_int = df.select_dtypes(include=('int64','float64'))
```

```python
cols_int = cols_int.columns
```

# Standart Scaler

```python
standartScaletData = standartScaling.fit_transform(df[cols_int])
```

```python
standartScaletData
```

```python
df_STDscaed_data = pd.DataFrame(standartScaletData, columns=cols_int, index=df.index)
```

```python
df_STDscaed_data
```

# Min Max Scaler

```python
min_maxScaled_data = minmaxScaling.fit_transform(df[cols_int])
```

```python
df_minmaxScaled = pd.DataFrame(min_maxScaled_data, columns=cols_int, index=df.index)
df_minmaxScaled
```

```python
df[cols_int] = robastScaling.fit_transform(df[cols_int])
```

```python
df.head()
```

```python
for cols in df.columns:
    if df[cols].isnull().any():
       if df[cols].dtype() == 'object':
           df[cols].fillna(df[cols].mode()[0], inplace=True)
       else:
           df[cols].fillna(df[cols].mean(), inplace=True)
```

```python
df
```

```python
for colmns in df.columns:
    if df[colmns] == 'object':
        if df[colmns].nunique() >= 5:
            dummies = pd.get_dummies(df[colmns],prefix='hot_encode' ,dtype=int)
            df = pd.concat([df.drop(columns=[colmns]), dummies], axis=1)
        else:
            df[colmns] = labelEncoding.fit_transform(df[colmns])
            
```
